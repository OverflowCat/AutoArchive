```
import json
import tweepy

auth = tweepy.OAuthHandler(T1, T2)
auth.set_access_token(T3, T4)
api = tweepy.API(auth)
```

```
User = "@OverflowCat" #758295043337617408
# Cursor is the search method this search query will return 20 of the users latest favourites just like the php api you referenced
favs = tweepy.Cursor(api.favorites, id=User).items(99)
urls = []
txts = []
for favorite in favs:
    # print('\n\n\nTweet Author:')
    # print('Screen Name: '+ favorite.user.screen_name)
    # print('Name: '+ favorite.user.name)
    # print('\nTweet:')
    # print('Tweet Id: ' + str(favorite.id))
    # print('Tweet Text: '+ favorite.text)
  urls.append("http://twitter.com/" + favorite.user.screen_name + "/status/" + str(favorite.id))
  txts.append(favorite.text)
```

```
print(urls)
```

```
arc_urls = []
for url in urls:
  arc_url = savepagenow.capture_or_cache(url)
  arc_urls.append(arc_url)
  sleep(5)
```

The limit of an IP is **15 times per minute**. That is to say we'd better sleep for 5 secs.
